```python
#seperating input and output features
X=df.drop(labels=['Id','Species'],axis=1)
Y=df[['Species']]
```

```python
X
```

```python
#feature scaling - Features in same range
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler().set_output(transform='pandas')
X_pre= scaler.fit_transform(X) #fitting the data X for scaling
```

```python
import pickle #pickle is used as a bridge to connect frontend and backend
with open('scaler.pkl','wb') as file1:
    pickle.dump(scaler,file1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.33,random_state=21)
```
